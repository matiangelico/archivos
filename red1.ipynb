# **IMPORTAMOS DATOS DE FUNCION f(x) = 1 / x**


---



```python
import pandas as pd

url = 'https://raw.githubusercontent.com/matiangelico/archivos/main/datos.csv'
datos = pd.read_csv(url,sep=",")

datos.head()
```

```python
X = datos[['x']]
y = datos['f(x)']

model = Sequential()
model.add(Dense(512, input_dim=1, activation='relu'))    
model.add(Dense(1, activation='linear'))    
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])
model.summary()
result = model.fit(X, y, validation_split=0.2, epochs=150, verbose=1)

```

```python
xt = np.array([12.8])
predictions = model.predict(xt)
predicted_value = np.array(predictions)
print(predicted_value)
```
